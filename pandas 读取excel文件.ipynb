# pandas 读取excel文件

pd.read_excel(io,sheet_name,header,index_col...)

- io=文件路径
- sheet_name=int/str/list/None
    - 读取单个sheet，sheet_name默认为0，返回第1个sheet，1则返回第1个sheet，2则返回第2个sheet，...类推；
    - 读取单个sheet，sheet_name=string, 如‘Sheet1’,‘Sheet2’,‘具体sheet名称’...
    - 读取多个sheet，sheet_name=list,如[0,1,2,5...'sheet名称'...]
    - 读取全部sheet，sheet_name=None
- header=int/None 列索引  
    - 默认0，表示第1行作为列名（列索引），1表示第二行作为列名（列索引）...类推；
    - None系统自动添加0,1,2...作为列索引；
- index_col=None/int/str
    - 默认None，它的行索引默认是 0,1,2,3…递增的数字；
    - 0表示指定第1列作为行索引，1表示指定第2列作为行索引，...类推；
    - index_col='列索引'

```python
import pandas as pd
df=pd.read_excel('pyexcel\\2017年级.xlsx',sheet_name=0,header=0,index_col=None)
df
```

```python
df.fillna(value='ok',inplace = True)
df
```
